```python
#!/usr/bin/python3.8
# Amazon share price
%matplotlib inline
import pandas as pd
import mplfinance as mpf
```

```python
file = 'AMZN.csv'
data = pd.read_csv(file)
```

```python
data
```

```python
data.info()
```

```python
data.columns
```

```python
data.Date = pd.to_datetime(data.Date)
data.info()
```

```python
data = data.set_index('Date')
```

```python
data
```

```python
mpf.plot(data)
```

```python
mpf.plot(data,type='line',volume=True)
```

```python
mpf.plot(data['2020-03'],volume=True)
```

```python
mpf.plot(data['2020-03'],type = 'candle',volume=True)
```

```python
mpf.plot(data['2020-03':'2020-07'],type = 'candle',mav=(20),volume=True)
```

```python
mpf.plot(data['2020-03':'2020-07'],type = 'candle',mav=(20),volume=True,tight_layout = True)
```

```python
mpf.plot(data['2020-03':'2020-07'],figratio=(20,12),
         type = 'candle',
         title = 'Amazon Price 2019/2020',
         mav=(20),volume=True,
         tight_layout = True)
```

```python
mpf.plot(data['2020-03':'2020-07'],figratio=(20,12),
         type = 'candle',
         title = 'Amazon Price 2019/2020',
         mav=(20),volume=True,
         tight_layout = True,
         style = 'yahoo')
```
